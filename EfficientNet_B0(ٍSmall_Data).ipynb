```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D

base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # تجميد الأوزان

# إنشاء النموذج الجديد
x = GlobalAveragePooling2D()(base_model.output)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)  # تقليل فرط التكيّف
output = Dense(1, activation='sigmoid')(x)  # طبقة الإخراج للتصنيف الثنائي

model = Model(inputs=base_model.input, outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

# Define the paths to your original dataset and the new train/val directories
benign_dir = "/content/benign"
malignant_dir = "/content/malignant"
train_dir = "/content/train"
val_dir = "/content/val"

# Create the train and val directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Create subdirectories for benign and malignant classes in train and val directories
os.makedirs(os.path.join(train_dir, "benign"), exist_ok=True)
os.makedirs(os.path.join(train_dir, "malignant"), exist_ok=True)
os.makedirs(os.path.join(val_dir, "benign"), exist_ok=True)
os.makedirs(os.path.join(val_dir, "malignant"), exist_ok=True)

# Split and copy images for the benign class
benign_images = os.listdir(benign_dir)
train_benign, val_benign = train_test_split(benign_images, test_size=0.2, random_state=42)
for image in train_benign:
    shutil.copy(os.path.join(benign_dir, image), os.path.join(train_dir, "benign", image))
for image in val_benign:
    shutil.copy(os.path.join(benign_dir, image), os.path.join(val_dir, "benign", image))

# Split and copy images for the malignant class
malignant_images = os.listdir(malignant_dir)
train_malignant, val_malignant = train_test_split(malignant_images, test_size=0.2, random_state=42)
for image in train_malignant:
    shutil.copy(os.path.join(malignant_dir, image), os.path.join(train_dir, "malignant", image))
for image in val_malignant:
    shutil.copy(os.path.join(malignant_dir, image), os.path.join(val_dir, "malignant", image))

print("Train and validation data created successfully!")
```
